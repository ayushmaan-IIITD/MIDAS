# b) Reading the JSONlines file and extracting important information

### Creating an empty dataFrame for displaying the data in tabular format

```python
data = pd.DataFrame(columns=['Text of Tweet','Date and Time','Number of Favorites','Number of Retweets','Number of Images'])
```

### Reading the tweets from jsonlines file and extracting important information

```python
with jsonlines.open('tweets.jsonl') as reader:
    i = 0
    for obj in reader.iter(type=dict, skip_invalid=True):
        data_list = []
        data_list.append(obj['full_text'])                          # extracting the text of the tweet
        data_list.append(obj['created_at'])                         # extracting the date and time of the tweet
        data_list.append(obj['favorite_count'])                     # extracting the number of likes/favorites of the tweet
        data_list.append(obj['retweet_count'])                      # extracting the number of retweets of the tweet
        if 'media' in obj['entities'].keys():                       # Checking if a tweet contains any attached media
            media_list = obj['extended_entities']['media']          # Extracting the list of attached media in the tweet
            image_count = 0
            for j in range(len(media_list)):
                if media_list[j]['type'] == 'photo':                # Checking if an attached media is of type 'photo' or not
                    image_count+=1
            if image_count == 0:
                data_list.append("None")
            else:
                data_list.append(image_count)
        else:
            data_list.append("None")
        data.loc[i] = copy.deepcopy(data_list)
        i+=1
```
